### 3. Correlation Matrix Heatmap

```python
# Create a correlation matrix using pandas

df_spotify_quantitative.corr()
```

```python

# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(11, 11)) # figure size 
plt.matshow(df_spotify_quantitative.corr(), fignum=f.number) # type of plot
plt.xticks(range(df_spotify_quantitative.shape[1]), df_spotify_quantitative.columns, fontsize=8, rotation=58) # x axis labels
plt.yticks(range(df_spotify_quantitative.shape[1]), df_spotify_quantitative.columns, fontsize=8) # y axis labels

cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=8) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title


plt.show()
```

### 4. Creating scatter-plots

##### 4.1 Duration - Number of artists

```python
# Create a scatter plot
plt.scatter(df_spotify_quantitative['artists_num'], df_spotify_quantitative['duration'], color ='orange')

# Add labels and title
plt.xlabel('artists_num')
plt.ylabel('duration')
plt.title('Duration and Number of Artists')

```

Since this is a mild correlation we can't fully assume that there is a strong relationship. 
It is surprising that there are so many songs with more than 5 artists. There seems to be an inflexion point just below 7.5 artists per song.

##### 4.2 Acousticness - Energy

```python
# Create a scatter plot
plt.scatter(df_spotify_quantitative['acousticness'], df_spotify_quantitative['energy'], color = 'darkred')

# Add labels and title
plt.xlabel('acousticness')
plt.ylabel('energy')
plt.title('Acousticness - Energy')
```

The more acoustic a song, the less energy it will have according to the spotify algorithm. 
Production wise this makes sense since EDM and hip hop are genres that have a very low acousticness and the best practices in the production are to make the songs as loud as possible.

These genres are also considered to be high energy genres, more specifically EDM.

The process of making a song loud happens in the mastering stage of a song. There are different mastering thresholds of loudness for different genres. 

***It would be very interesting to see the loudest genres from this dataset!!!**

As a music producer it would also be interesting to know other loudness averages for other genres. 

##### 4.3 Peak rank - Chart weeks

```python
# Create a scatter plot
plt.scatter(df_spotify_quantitative['peak_rank'], df_spotify_quantitative['chart_weeks'], color = 'darkred')

# Add labels and title
plt.xlabel('peak_rank')
plt.ylabel('chart_weeks')
plt.title('Peak rank - Chart weeks')
```

Watching the "white space" of the chart, tells us that there are no songs that had a peak rank 175 and lasted more than 150 weeks in the chart. 
A song with a great peak rank such as 10, will slowly descend the charts having more time overall in the charts compared to songs that rank 175 or 200. 
Some songs may reach the charts at rank 200 and never climb any other spot. 
This is one hint that a song will most likely achieve their highest spot in its earliest time, but there is no outstanding proof of this other than intuition and this chart.

### 6. Categorical Plot: Peak rank - chart weeks

```python
# Create groups/categories

df_spotify_quantitative.loc[df_spotify_quantitative['peak_rank'] < 50, 'rank_tier'] = 'Top'
df_spotify_quantitative.loc[(df_spotify_quantitative['peak_rank'] >= 50) & (df_spotify_quantitative['peak_rank'] < 150), 'rank_tier'] = 'Mid'
df_spotify_quantitative.loc[df_spotify_quantitative['peak_rank'] >= 150, 'rank_tier'] = 'Bottom'
```

```python
# Set up the figure and axis
plt.figure(figsize=(10, 6))

# Plot using bar plot
for tier in df_spotify_quantitative['rank_tier'].unique():
    subset = df_spotify_quantitative[df_spotify_quantitative['rank_tier'] == tier]
    plt.scatter(subset['peak_rank'], subset['chart_weeks'], label=tier)

# Add labels and title
plt.xlabel('Peak Rank')
plt.ylabel('Chart Weeks')
plt.title('Peak Rank vs. Chart Weeks')
plt.legend(title='Rank Tier')

plt.show()
```
